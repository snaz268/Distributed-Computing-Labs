```python
# generate random integer values
from random import seed
from random import randint
# seed random number generator
seed(1)
# A single node of a singly linked list
class Node:
  # constructor
    def __init__(self, data = None, next=None): 
        self.data = data
        self.next = next

# A Linked List class with a single head node
class LinkedList:
    def __init__(self):  
        self.head = None
  
  # insertion method for the linked list
    def create_ring(self, data_list):
        #For each node in data_list, create a new node and connect it in linked list
        for node_id in data_list:
            #Creating new node and connecting
            newNode = Node(node_id)
            
            if(self.head):
                current = self.head
                while(current.next):
                    current = current.next
                current.next = newNode
            else:
                self.head = newNode
                
        #Connecting the last node to 1st node      
        current = current.next
        current.next = self.head  
  
    def traverse_asc(self):
        count = 0
        current = self.head
        node_id = 0
        while(current):
            if(node_id == current.data):
                print("Leader = ", node_id)
                print("Number of steps taken to find leader = ", count)
                print("\n")
                return
            else:
                node_id = max(current.data, node_id)
                print("current.data = ",current.data)
                print("Max node_id  = ",node_id)
                #print("\n")
                current = current.next
                count += 1

            
    def traverse_desc(self):
        count = 0
        current = self.head
        node_id = 100
        while(current):
            if(node_id == current.data):
                print("Leader = ", node_id)
                print("Number of steps taken to find leader = ", count)
                return
            else:
                node_id = min(current.data, node_id)
                print("current.data = ",current.data)
                print("Max node_id  = ",node_id)
                #print("\n")
                current = current.next
                count += 1

# Singly Linked List with insertion and print methods
LL = LinkedList()
data = [1,2,3,4]
rand_list = []

LL.create_ring(data)
print("Data List = ", data)
print("*** Ascending Order Election ***")
LL.traverse_asc()
print("*** Descending Order Election ***")
LL.traverse_desc()


print("\n")
print("*** Generating Random List ***")
for i in range(4):
    value = randint(0, 10)
    rand_list.append(value)
print("Random Data List = ", rand_list)
print("\n")

print("*** Ascending Order Election ***")
LL.traverse_asc()
print("*** Descending Order Election ***")
LL.traverse_desc()
```
